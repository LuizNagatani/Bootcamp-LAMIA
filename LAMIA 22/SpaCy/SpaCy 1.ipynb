```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
with open('wiki_us.txt', 'r') as f:
    text = f.read()

text
```

```python
doc = nlp(text)
```

```python
doc
```

```python
print(len(text))
print(len(doc))
# o arquivo txt foi agrupado em palavras ao invés de caracteres
```

```python
for token in text[:10]:
    print(token)
```

```python
for token in doc[:10]:
    print(token)
```

```python
contagem = 0
for sent in doc.sents:
  print(contagem)
  print(sent)
  contagem +=1
```

```python
sentence1 = list(doc.sents)[0] # obtem a primeira frase, no caso a "The united States of America..."
sentence1
```

```python
token2 = sentence1[2] #ao selecionar um determinado indice de uma frase, se obtem a palavra correspondente ao índice
token2
```

```python
token2.left_edge # obtem a palavra mais a esquerda desde que ela seja descendente sintática
```

```python
token2.right_edge # realiza a mesma operação só que para a direita
```

```python
token2.ent_type_ #identifica o tipo do token, no caso a palavra é identificada como uma descrição de espaço geopolítico
```

```python
token2.ent_iob_
```

```python
token2.lemma_ #obtem a palavra sem ser conjugada, util para verbos
```

```python
token2.morph # realiza uma análise morfológica da palavra, no caso este token é um substantivo singular
```

```python
text = "Mike enjoys playing football."
doc2 = nlp(text)
print(doc2)
```

```python
for token in doc2:
    print(token.text, token.pos_, token.dep_) #expõe a palavra, classe gramatical e sua dependencia sintática
```

```python
from spacy import displacy
displacy.render(doc2, style='dep')
```
