```python
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score

X = pd.read_csv('entradas_breast.csv')
y = pd.read_csv('saidas_breast.csv')
```

```python
def criaRede():
    # creating neurons 
    clas = Sequential()
    clas.add(Dense(units=16, activation='relu', 
                            kernel_initializer='random_uniform', input_dim = 30))
    clas.add(Dropout(0.4))
    clas.add(Dense(units=16, activation='relu', 
                            kernel_initializer='random_uniform'))
    clas.add(Dropout(0.4))
    clas.add(Dense(units=1, activation='sigmoid'))
  
    opm = keras.optimizers.Adam(learning_rate=0.001, weight_decay=0.0001, clipvalue=0.5)
    clas.compile(optimizer=opm, loss='binary_crossentropy',
                        metrics= ['binary_accuracy'])
    return clas
```

```python
clas = KerasClassifier(model = criaRede, epochs = 100, batch_size = 10)
clas
```

```python
res = cross_val_score(estimator = clas, X = X, y = y, cv =10, scoring = 'accuracy')
res
```

```python
print(f'Média: {res.mean()}  Erro: {res.std()}')
```
