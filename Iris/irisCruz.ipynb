```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
```

```python
base =pd.read_csv('iris.csv')
X = base.iloc[:, 0:4].values
y = base.iloc[:, 4].values
```

```python
label = LabelEncoder()
y = label.fit_transform(y)
y_dummy = to_categorical(y)
```

```python
def criaRede():
    clas = Sequential()
    clas.add(Dense(units=4, activation = 'relu', input_dim = 4))
    clas.add(Dense(units=4, activation = 'relu'))
    clas.add(Dense(units=3, activation = 'softmax'))
    clas.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
    return clas
```

```python
clas = KerasClassifier(build_fn = criaRede, epochs = 1000, batch_size = 10)
res = cross_val_score(estimator = clas, X = X, y = y_dummy, cv = 10, scoring = 'accuracy')
```

```python
params = {'batch_size' :[10, 30], 
          'epochs':[1000, 3000], 'optmizer': ['adam'], 
          'loss_function':['categorical_crossentropy'], 
          'kernel_initializer':['random_uniform', 'normal'],
          'activation' : ['relu'],
          'neurons' : [4, 8]}
grid_search = GridSearchCV(estimator = clas, param_grid = params, scoring = 'accuracy', cv = 5)
            
    
```

```python
media = res.mean()
desvio = res.std()
print(f'média = {media}, desvio = {desvio}')

```
