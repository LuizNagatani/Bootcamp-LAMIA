```python
import spacy
from spacy.matcher import Matcher
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
matcher = Matcher(nlp.vocab)
pattern = [
    {
        'LIKE_EMAIL': True
    }
]
matcher.add('EMAIL_ADDRESS', [pattern])
```

```python
doc = nlp('This is an email address: wmattingly@aol.com, this is an email too: mathally@yhoo.com')
matches = matcher(doc)
```

```python
matches
```

```python
for match in matches:
  print(match, doc[match[1]:match[2]]) #printa o token onde o email se inicia e encerra, seguido do email em si
```
