```python
import spacy
```

```python
nlp = spacy.load('en_core_web_sm')
text = 'West Chestertenfieldville was referenced in Mr. Deeds.'
```

```python
doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_) #assume que west Cherster... é um local e Deeds é uma pessoa por conta do contexto fornecido pelo título "Mr."
```

```python
ruler = nlp.add_pipe('entity_ruler')
```

```python
patterns = [
    {
        'label': 'GPE',
        'pattern':'Chestertenfieldville'
    }
]
```

```python
ruler.add_patterns(patterns)
```

```python
doc2 = nlp(text)
for ent in doc2.ents:
    print(ent.text, ent.label_)
```

```python
nlp2 = spacy.load('en_core_web_sm')
```

```python
ruler = nlp2.add_pipe('entity_ruler', before='ner')
```

```python
ruler.add_patterns(patterns)
```

```python
for ent in doc2.ents:
    print(ent.text, ent.label_) #este processo foi feito para reconhecer west chester... como uma GPE
```

```python
nlp3 = spacy.load('en_core_web_sm')
```

```python
ruler = nlp3.add_pipe('entity_ruler', before='ner')
```

```python
patterns = [
    {
        'label': 'GPE',
        'pattern':'Chestertenfieldville'
    },
    {
        'label': 'FILM',
        'pattern':'Mr. Deeds'
    }
]
```

```python
ruler.add_patterns(patterns)
```

```python
doc3 = nlp3(text)
for ent in doc3.ents:
    print(ent.text, ent.label_) #utilizando as informações apresentadas em patterns, o modelo parou de analisar deeds como uma pessoa, e agora lê Mr. Deeds como o nome de um filme
```
