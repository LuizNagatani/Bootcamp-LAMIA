```python
import spacy
from spacy.matcher import Matcher
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
matcher = Matcher(nlp.vocab)
pattern = [
    {'LIKE_NUM' : True}
]
matcher.add('NUMERO', [pattern])
```

```python
with open('Tboi.txt', 'r') as f:
    text = f.read()

text
```

```python
doc = nlp(text)
matches = matcher(doc)
doc
```

```python
matches
for match in matches:
    print(match, doc[match[1]:match[2]])
```

```python
itens =0
for match in matches:
    itens += int(doc[match[1]:match[2]].text)
itens
```
